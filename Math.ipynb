### Volume concentration in higher dimensions

Inorder to consider the Volume occupied by the outermost shell we consider a smaller sphere of radius (1-e) and the percentage of volume occupied by this thin outer shell will be of the form $1-(1-e)^d$, where d is the dimension and e is a tiny portion say 0.1. When the dimension is small less volume is occupied by the outershell but as dimension increases this changes...

```python
e = 0.1
d = np.arange(100)
for i in d:
    V = 1-(1-e)**d
print(V)
```

As dimension reaches 99 almost 99.9997 percentage volume is occupied by the outermost shell which tells us that as the dimension increases most of the volume is at the outermost shell and the inside is hollow. When trying to find datapoints in higher dimension most of the datapoints will be on the outermost shell and less datapoints can be seen inside. As d $\to \infty$ the inside of the shell will be hollow.

```python
e = 0.1
d = np.arange(100)
plt.figure(figsize=(10,5))
for i in d:
    V = 1-(1- e)**d #volume occupied in a outershell
    plt.plot(d,V,marker='.',linestyle=' ') 
plt.xlabel('Dimension')
plt.ylabel("Volume occupied by outershell")
plt.grid(alpha=0.5)
plt.title('Volume occupied by outershell in Higher Dimension')
```

### Volume of D-sphere

```python
r = 1
Volume = []
D = np.arange(30)
for i in range(30):
    V = (((np.pi)**(i/2))/math.gamma(i/2 + 1))*(r**i)
    Volume.append(V)
plt.figure(figsize=(8, 6))
plt.plot(D,Volume,marker='o',linestyle='--',color='gray')
plt.xlabel("Dimension")
plt.ylabel("Volume of Hypersphere")
plt.title('Volume of Hyphersphere vs Dimension')
plt.grid(alpha=0.5)
plt.show()
```

There is a gamma function in the denominator initially the $\pi^{n/2}$ term is higher than denominator but as the dimension goes up the factorial from the gamma function fires up and due to this the Volume goes down((mathematically))

More generalised code

```python
theta = np.linspace(0, 2*np.pi,100)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()
D = np.arange(1,9)
for n,i in enumerate(D):
    ax = axes[n]
    x = np.abs(np.sin(theta)**i)*np.sign(np.sin(theta)) #i = 2/p
    y = np.abs(np.cos(theta)**i)*np.sign(np.cos(theta))
    ax.plot(x,y,linestyle='-.')
    ax.set_title(f"Dimension {i+1}")
```

Well now things might get a bit confusing as we know in 2dimension we are supposed to get a circle and 3D a sphere but here instead of a sphere we are getting a diamond... wonder why??? well if we were to get a sphere we should have included a new coordinate along with the addition of a new angle but since we are not adding a coord but multiplying a new angle to x and y this happens.

This is all from the fact that $x^2$ + $y^2$ = 1 and from it we generalise the equation as |$x^p$| + |$y^p$| = 1 and this is the mathematical equation for a superelipse for which we get x = sin $\theta^{\frac{2}{p}}$ and y = cos $\theta^{\frac{2}{p}}$ and in the code we have i instead of the fraction there hence for i=1 we are getting a circle. 

This equation works as it relies on $sin^2 \theta$ + $cos^2 \theta$ = 1 to draw closed loops.

x = $(sin \theta)^i$ => $sin \theta$ = $x^{\frac{1}{i}}$  (From parametarization)

y = $(cos \theta)^i$ => $cos \theta$ = $y^{\frac{1}{i}}$

$\frac {dx}{d\theta}$ = i $\times$ $(sin \theta)^{i-1}$ $\times cos \theta$

$\frac {dy}{d\theta}$ = i $\times$ $(cos \theta)^{i-1}$ $\times sin \theta$

$\frac{dy}{dx}$ = -cot $\theta^{i-2}$

For $\theta $ = 30 ; cot(30) = $\sqrt{3} \approx 1.73$ => for $1.73^{i-2}$ as i goes up the slope becomes very steep

FOr $\theta $ = 60 ; cot(60) = $\frac{1}{\sqrt{3}} \approx 0.577$ for $0.577 ^{i-2}$ as i goes up the slope goes down

For $\theta$ > 45 the slope is very flat and for $\theta$ < 45 the slope is very steep so the resulting tangents would be either vertical or horizontal which shows the spiky nature.

so for circle we have $(sin \theta)^{\frac{2}{i}}$ and $(cos \theta)^{\frac{2}{i}}$ where i is 1 from the code so we get a circle, when i is 2 we get a diamond as it would just be |x| + |y| = 1 which would give us four slopes in each quadrant.

For i = 2 if x= 0.5 the y would be also be 0.5 so that they could add up to 1 but for i = 4 at x = 0.5 we have to take the square root so that we get 0.707 and $\sqrt{y}$ is 0.293 and the y axis would be 0.085 the point gets sucked in this is why as the dimension goes up the sphere gets more spiky and more sucked up.


Higher dimensional spheres have a relation : $|x_1|^p + |x_2|^p +.....+ |x_n|^p$ = 1 so as p = 1 it would be a diamond for p = 2 it would be a sphere and as p goes up it becomes a cube.

Distance from the origin to a point is found from : D = $\sqrt{{x_1}^2+{x_2}^2+...+{x_n}^2}$

Instead of square root($\frac{1}{2}$) we can generalize it to $\frac{1}{p}$ so that : D = $({{x_1}^2+{x_2}^2+...+{x_n}^2})^{\frac{1}{p}}$ and since we are considering unit sphere the value of D becomes 1 and we get the equation.

In 2D |x|+|y| = 1 would give us a diamond $x^2$ + $y^2$ = 1 would give us a sphere and when the value of p goes up we get a cube

Its just like normal coordinate geometry where a new variable pops up as dimension goes up $x^2$ + $y^2$ = 1 for circle , $x^2$ + $y^2$ + $z^2$ = 1 for sphere and it would be $w^2$ +$x^2$ + $y^2$ + $z^2$ = 1 for 4D, inorder to account for other points the sides get sucked up inside.

```python
theta = np.linspace(0, 2*np.pi,100)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()
D = np.linspace(1,2,8)
for n,i in enumerate(D):
    ax = axes[n]
    x = np.sign(np.sin(theta))*np.abs(np.sin(theta))**i
    y = np.sign(np.cos(theta))*np.abs(np.cos(theta))**i
    ax.plot(x,y,linestyle='-.',color='black')
```

In this we can clearly see how the surface is coming to look like a diamond as the value for i changes

### Shift in Peak with changing radii

```python
fig, axes = plt.subplots(4,2,figsize=(20,20))
axes = axes.flatten()
D = np.arange(1,25)

for n, r in enumerate(range(1,9)):
    ax = axes[n]
    Volume = []
    for d in D:
        d = int(d) #numpy integer cant go above certain power so it would go negative  
        V = (np.pi**(d/2)/math.gamma(d/2+1))*r**d
        Volume.append(V)

    ax.plot(D, Volume, marker='o',linestyle="--",color='gray')
    ax.set_title(f"Radius of hypersphere = {r}")
    ax.set_xlabel("Dimension")
    ax.set_ylabel("Volume")
    ax.grid(True,alpha=0.5)

plt.tight_layout()
plt.show()
```

### Finding the Curvature Value at 45 degree

```python
theta = np.deg2rad(np.linspace(0,90,100))
fig,axes = plt.subplots(2,4,figsize=(20,10))
axes = axes.flatten()
def line(x,m):
    return x*m
D = np.arange(1,9)
for n,i in enumerate(D):
    ax = axes[n]
    x = np.sign(np.sin(theta))*np.abs(np.sin(theta))**i
    y = np.sign(np.cos(theta))*np.abs(np.cos(theta))**i
    ax.plot(x,y,linestyle='-.',color='black')
    max_len = np.sin(np.deg2rad(45))**i
    m =1
    a = np.linspace(0,max_len,100)
    b = line(a,m) 
    ax.plot(a,b,linestyle='--',color='gray')
    ax.grid(True,alpha=0.3)
    ax.set_title(f"Dimension {i+1}")
    print("(",max_len,",",max_len,")",f"are the points at curvature for dimension {i+1}")
```

### Seperated View of d-Sphere

```python
rho = 1
theta = np.linspace(0,2*np.pi,100)
phi = np.linspace(0,np.pi,100)

T,P = np.meshgrid(theta,phi)

fig,axes = plt.subplots(1,2,figsize=(15,10))
axes = axes.flatten()

r = rho*np.sin(P)
x = r*np.cos(T)
y = r*np.sin(T)
z = rho*np.cos(P)
axes[0].plot(x, y, color="black", alpha=0.3)
axes[0].set_title("XY Projection (Top View)")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].set_aspect("equal")

axes[1].plot(x, z, color="darkgray", alpha=0.3)
axes[1].set_title("XZ Projection (Front View)")
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Z-axis")
axes[1].set_aspect("equal")

plt.tight_layout()
plt.show()
    
```
