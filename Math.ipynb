### Volume concentration in higher dimensions

Inorder to consider the Volume occupied by the outermost shell we consider a smaller sphere of radius (1-e) and the percentage of volume occupied by this thin outer shell will be of the form $1-(1-e)^d$, where d is the dimension and e is a tiny portion say 0.1. When the dimension is small less volume is occupied by the outershell but as dimension increases this changes...

```python
e = 0.1
d = np.arange(100)
for i in d:
    V = 1-(1-e)**d
print(V)
```

As dimension reaches 99 almost 99.9997 percentage volume is occupied by the outermost shell which tells us that as the dimension increases most of the volume is at the outermost shell and the inside is hollow. When trying to find datapoints in higher dimension most of the datapoints will be on the outermost shell and less datapoints can be seen inside. As d $\to \infty$ the inside of the shell will be hollow.

```python
e = 0.1
d = np.arange(100)
plt.figure(figsize=(10,5))
for i in d:
    V = 1-(1- e)**d #volume occupied in a outershell
    plt.plot(d,V,marker='.',linestyle=' ') 
plt.xlabel('Dimension')
plt.ylabel("Volume occupied by outershell")
plt.grid(alpha=0.5)
plt.title('Volume occupied by outershell in Higher Dimension')
```

#### Imagining a Hypercube

Now the task is to think how it would look for different dimensions, we know how a square and cube looks like but what about a 3D hypercube or higher dimension think of it like a square but just the diagonals being pulled while the side lengths remain the same so when dimension goes up as we saw above the diagonal length increases as well.

To do this visualisation consider each point that would be needed to imagine this so since the sides are unit length and the diagonals are changing we can consider each points (0.5,0),(d,d),(0,0.5),(-d,d),..... and so on till (d,-d),(0.5,0) and then we come all the way around.

```python
plt.figure(figsize=(7,7))
D = np.arange(1,10)
for i in D:
    d = np.sqrt(i)/2
    x_corners = [0.5,d,0,-d,-0.5,-d,0,d,0.5]
    y_corners = [0,d,0.5,d,0,-d,-0.5,-d,0]
    plt.plot(x_corners,y_corners,linestyle='-.',label=f'Dim {i}')
plt.legend(loc=9,ncols=3)
plt.title('Cube geometry in Higher dimensions')
```

Well what I had expected was not like this but it should have looked like a square with just the sides being pulled more and more as the dimension goes up thats how it looks, for imagining a 4D cube think of the tesseract from Avengers Movie a cube inside a cube

### Volume of D-sphere

```python
r = 1
Volume = []
D = np.arange(30)
for i in range(30):
    V = (((np.pi)**(i/2))/math.gamma(i/2 + 1))*(r**i)
    Volume.append(V)
plt.figure(figsize=(8, 6))
plt.plot(D,Volume,marker='o',linestyle='--',color='gray')
plt.xlabel("Dimension")
plt.ylabel("Volume of Hypersphere")
plt.title('Volume of Hyphersphere vs Dimension')
plt.grid(alpha=0.5)
plt.show()
```

There is a gamma function in the denominator initially the $\pi^{n/2}$ term is higher than denominator but as the dimension goes up the factorial from the gamma function fires up and due to this the Volume goes down((mathematically))

#### Showing how hypershere becomes spiky as dimension goes up

```python
theta = np.linspace(0, 2*np.pi,100)
x = np.sin(theta)**3
y = np.cos(theta)**3
plt.plot(x,y)
```

right when I tried with even dimension I understood due to the sign becoming positive the diagram was not looking like how it is supposed to,so I found this function called np.sign that gives the sign value for that particular theta value and got the spiky shape back

```python
theta = np.linspace(0, 2*np.pi,100)
x = np.abs(np.sin(theta)**4)*np.sign(np.sin(theta))
y = np.abs(np.cos(theta)**4)*np.sign(np.cos(theta))
plt.plot(x,y)
```

More generalised code

```python
theta = np.linspace(0, 2*np.pi,100)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()
D = np.arange(1,9)
for n,i in enumerate(D):
    ax = axes[n]
    x = np.abs(np.sin(theta)**i)*np.sign(np.sin(theta))
    y = np.abs(np.cos(theta)**i)*np.sign(np.cos(theta))
    ax.plot(x,y,linestyle='-.')
    ax.set_title(f"Dimension {i}")
```
